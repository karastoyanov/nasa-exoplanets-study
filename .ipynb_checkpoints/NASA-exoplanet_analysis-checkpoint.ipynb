```python
import pandas as pd
import numpy as np
spreadsheet = pd.read_csv('src/nasa_exoplanets.csv', delimiter = ',', low_memory = False)
spreadsheet.head()
```

```python
df = pd.DataFrame(spreadsheet)
years = []
for year in df.disc_year:
    if year not in years:
        years.append(int(year))
    years.sort()
print(f'Years of exoplanets discovered: {",".join([str(x) for x in years])}')


for year in years:
    current_year_found = df.query(f"disc_year == '{year}'")
    print(f'Total Exoplanets found in {year}: {current_year_found}')
    
```

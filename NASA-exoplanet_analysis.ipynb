### NASA's search for another earth.

#### Current project aims to give insights on the search of another life in our Universe and the method used by NASA in discovering another worlds out there.

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read CSV file from local source
spreadsheet = pd.read_csv('src/nasa_exoplanets.csv', delimiter = ',', low_memory = False)
spreadsheet.head()
```

```python
df = pd.DataFrame(spreadsheet)
years = []
for year in df.disc_year:
    if year not in years:
        years.append(int(year))
    years.sort()
print(f'Years of discoveries: {",".join([str(x) for x in years])}')

planets_by_year = {"Year": "Number"}
for year in range(len(years)):
    current_year = str(years[year])
    number_planets_found = df.query(f"disc_year == '{current_year}'")
    planets_by_year[current_year] = len(number_planets_found)
    print(f'Total number of exoplanets found in {current_year}: {len(number_planets_found)}')
```

```python
courses = list(planets_by_year.keys())
values = list(planets_by_year.values())
fig = plt.figure(figsize = (80, 15))
courses = courses[1:-1]
values = values[1:-1]

plt.pie(values, labels = ([str(x) for x in courses]))
plt.show()
```
